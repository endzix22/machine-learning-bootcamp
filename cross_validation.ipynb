Walidacja krzy≈ºowa

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
import plotly.express as px

np.random.seed(42)
sns.set(font_scale=1.3)
```

Wygenerowanie danych

```python
raw_data = make_moons(n_samples=700, noise=0.25, random_state=42)
data = raw_data[0]
target = raw_data[1]

print(f'{data[:5]}\n')
print(f'{target[:5]}')
```

```python
df = pd.DataFrame(data=np.c_[data, target], columns=['x1', 'x2', 'target'])
df.head()
```

Wizualizacja danych

```python
px.scatter(df, x='x1', y='x2', color='target', width=700, height=400)
```

Podzia≈Ç na zbi√≥r treningowy i testowy

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target)

print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

Wizualizacja zbioru treningowego i testowego

```python
plt.figure(figsize=(10, 8))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='RdYlBu', label = 'training_set')
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='RdYlBu', marker='x', alpha=0.5, label='test_set')
plt.title('Zbi√≥r treningowy i testowy')
plt.legend()
plt.show()
```

Budowa modelu

```python
#drzewa decyzyjne 

from sklearn.tree import DecisionTreeClassifier
from mlxtend.plotting import plot_decision_regions

classifier=DecisionTreeClassifier(max_depth=5,min_samples_split=5)
classifier.fit(X_train,y_train)

plt.figure(figsize=(10,8))
plot_decision_regions(X_train,y_train,classifier)
plt.title(f'Zbior treningowy: dokladnosc {classifier.score(X_train,y_train):.4f}')
plt.show()
```

```python

plt.figure(figsize=(10,8))
plot_decision_regions(X_test,y_test,classifier)
plt.title(f'Zbior treningowy: dokladnosc {classifier.score(X_test,y_test):.4f}')
plt.show()
```

WALIDACJA KRZY≈ªOWA

```python
from sklearn.model_selection import cross_val_score
#do drzewa decyzyjnego
classifier = DecisionTreeClassifier(max_depth=5, min_samples_split=5)
#cross_val_score
scores = cross_val_score(estimator=classifier, X=X_train, y=y_train, cv=10)
scores
#do zmiennej scores rzypisane wyniki walidacji krzyzowej 
#estimator=nasz_model
#X=dane treningowe, Y=dane treningowe
#cv=liczba podzialow walidacji
```

```python
#zwrocona lista 10 elemenetow zawierajaca dokladnosc
```

```python
#teraz wezmiemy sie za usrednienie tych wynikow

print(f'Accuracy: {scores.mean():.4f} (+/- {scores.std():.4f})')
#srednia z 10 accuracy =koncowy wynik
#std 
```

```python
classifier = DecisionTreeClassifier(max_depth=5, min_samples_split=5)
#cv zmienione na 15 i obserwujemy wyniki 
scores = cross_val_score(estimator=classifier, X=X_train, y=y_train, cv=15)
scores
```

```python
print(f'Accuracy: {scores.mean():.4f} (+/- {scores.std():.4f})')
```

```python
#z wynikow utworzmy DF by moc je wykreslic
```

```python
scores = pd.DataFrame(scores, columns=['accuracy'])
scores
```

wyniki z 15 accuracy

```python
px.bar(scores, y='accuracy', color='accuracy', width=700, height=400, 
       title=f'Walidacja krzy≈ºowa (15 podzia≈Ç√≥w) | Accuracy: {scores.mean()[0]:.4f} (+/- {scores.std()[0]:.3f})',
       color_continuous_scale=px.colors.sequential.Inferno_r, range_color=[scores.min()[0] - 0.01, 1.0])
```
