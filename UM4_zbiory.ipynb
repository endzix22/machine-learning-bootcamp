```python
import numpy as np 
import pandas as pd 
import sklearn 
import seaborn as sns 
```

```python
#ustawienia do wyswietlen....
np.random.seed(42)
np.set_printoptions(edgeitems=30, linewidth=100000, 
                  formatter=dict(float=lambda x: f'{x:.2f}'))

#ustawienie biblioteki seaborn jako domyslnej 
sns.set()
```

Załadowanie danych

```python
#skopiowanie danych z tej biblioteki ....powszechnie dostepnych
from sklearn.datasets import load_iris 
raw_data=load_iris()
#zbior z kwiatami 'iris.data', jest on w formie slownika
raw_data_copy=raw_data.copy()
#kopia danych
raw_data_copy.keys()
#wysiwetlimy klucze tego slownika
```

```python
print(raw_data_copy['DESCR'])
#opis danych
```

Przygotowanie danych

```python
## ogolnie zbior data i target 
data = raw_data_copy['data']
target = raw_data_copy['target']

#wyswietlimy 5 pierwszych wierszy 
print(f'{data[:5]}\n')
#wyswietlimy 5 pierwszych elementow
print(target[:5])
```

```python
#polaczenie atrybutow ze zmienna docelowa
all_data=np.c_[data,target]
#head jest w pandas
all_data[:5]
```

```python
#budowa obiektu DataFrame 
df=pd.DataFrame(data=all_data,columns=raw_data.feature_names+['target'])
df.head()
```

```python
data=df.copy()
#chcemy wyodrebnic kolumne target 
target=data.pop('target')
data.head()
#data juz jest bez target 
```

```python
target.head()
```

PODZIAŁ NA ZBIÓR TRENINGOWY I TESTOWY-iris data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)

#posprawdzamy sobie jak to wyglado ilosciowo
print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
#ile % do testu
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
#ilosc klas +ilosc ...
print(f'\ny_train:\n{y_train.value_counts()}')
print(f'\ny_test:\n{y_test.value_counts()}')
```

```python
#zmienienie na 30% z 25% ilosc danych w zbiorze testowym
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.3, random_state=42)

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
print(f'\ny_train:\n{y_train.value_counts()}')
print(f'\ny_test:\n{y_test.value_counts()}')
```

```python
#zmienie ilosci danych w zbiorze treningowym na 90%
X_train, X_test, y_train, y_test = train_test_split(data, target, train_size=0.9, random_state=42)

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
print(f'\ny_train:\n{y_train.value_counts()}')
print(f'\ny_test:\n{y_test.value_counts()}')
```

OTrzymanie rownego stosunku zmiennych w klasach dla obu zbiorow, dzieki:
stratify=target

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42, test_size=0.1, stratify=target)

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
print(f'\ny_train:\n{y_train.value_counts()}')
print(f'\ny_test:\n{y_test.value_counts()}')
```

```python
#rozdzielenie danych
data=raw_data_copy['data']
target=raw_data_copy['target']

print(f'{data[:5]}\n')
print(target[:5])
```

```python
target.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=40, test_size=0.25)

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
print(f'\ntarget:\n{target.value_counts() / len(target)}')
print(f'\ny_train:\n{y_train.value_counts() / len(y_train)}')
print(f'\ny_test:\n{y_test.value_counts() / len(y_test)}')
```

ROWNY PODZIAL ZE WZGLEDU NA ZMIENNA DOCELOWA

stratify=target

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=40, test_size=0.25, stratify=target)

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
print(f'\ntarget:\n{target.value_counts() / len(target)}')
print(f'\ny_train:\n{y_train.value_counts() / len(y_train)}')
print(f'\ny_test:\n{y_test.value_counts() / len(y_test)}')
```
