````python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns

#ustawienia formatowania
np.random.seed(42)
np.set_printoptions(precision=6, suppress=True, edgeitems=30, linewidth=120, 
                    formatter=dict(float=lambda x: f'{x:.2f}'))
sns.set(font_scale=1.3)
````

Wygenerowanie danych

````python
#do modelu regresyjnego, mozna to zrobic przy pomocy biblioteki:
#do cwiczen juz gotowa opcja
from sklearn.datasets import make_regression

data, target = make_regression(n_samples=100, n_features=1, n_targets=1, noise=30.0, random_state=42)
#fajnie bo od razu w gotowcu mamy podzileone dane 
#nie musimy tego robic sami 
print(f'data shape: {data.shape}')
print(f'target shape: {target.shape}')
````


### <a name='2'></a> Regresja Liniowa przy uÅ¼yciu scikit-learn

$Y = w_0 + w_1X_1$


````python
from sklearn.linear_model import LinearRegression 

regressor=LinearRegression()

#metoda fit() dopasowuje model liniowy do danych 
regressor.fit(data,target)
````

````python
#metoda score() dokonuje oceny modelu na przekazanych danych (wynik R2 score)
#czyli ocena jak nasz mdoel zostal dopasowany przez fit()

regressor.score(data,target)
#wynik im blizej 1 tym lepiej... to oznacza ze jest dopasowany do naszych danych
````

````python
#nasz model juz jest wytrenowny przy pomocy fit() i teraz dokonujemy predykcji 
y_pred=regressor.predict(data)
y_pred
#jakby nasz model dopasowal dane a i b, na podsatwie tego zostal policzony y_pred
````

Wizualizacja graficzna modelu

````python
plt.figure(figsize=(8,6))
plt.title('Regresja liniowa')
plt.xlabel('cecha x:')
plt.ylabel('zmienna docelowa ')
plt.scatter(data,target,label='cecha x')
#punktowo
plt.plot(data,y_pred,color='red',label='model')
plt.legend()
plt.show()
````

### <a name='4'></a> WspÃ³Åczynnik $R^{2}$ 
WspÃ³Åczynnik $R^2$ jest zdefiniowany jako $1 - \frac{u}{v}$, gdzie $u$ jest okreÅlone wzorem 
```
((y_true - y_pred) ** 2).sum()
``` 
oraz $v$ jako 
```
((y_true - y_true.mean()) ** 2).sum()
``` 
Czyli 
```
1 - (((y_true - y_pred) ** 2).sum()) / (((y_true - y_true.mean()) ** 2).sum())
```
NajlepszÄ moÅ¼liwÄ wartoÅciÄ jest 1.0. $R^{2}$ moÅ¼e przyjmowaÄ wartoÅÄ ujemnÄ. Model, ktÃ³ry przewiduje zawsze dla kaÅ¼dej prÃ³bki wartoÅÄ oczekiwanÄ zmiennej docelowej ma wspÃ³Åczynnik $R^2$ rÃ³wny 0.

````python
regressor.score(data,target)
````

### <a name='5'></a> KoÅcowa postaÄ modelu

$$Y = w_0 + w_1X_1$$
PostaÄ modelu: $$Y = 3.495 + 49.83 \cdot X_1$$

````python
plt.figure(figsize=(8,6))
plt.title('Regresja liniowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(data,target,label='cecha x')
plt.plot(data,regressor.intercept_+regressor.coef_[0]*data,color='red',label='model')
plt.legend()
plt.show()
````

#REGRESJA Z PODZIAÅEM NA ZBIÃR TRENINGOWY I TESTOWY

````python
#pobierzemy gotowe dane 

data, target = make_regression(n_samples=1000, n_features=1, n_targets=1, noise=15.0, random_state=42)

print(f'data shape: {data.shape}')
print(f'target shape: {target.shape}')
````

````python
#dzielimy na zbior treningowy i testowy 

from sklearn.model_selection import train_test_split 
X_train, X_test,y_train,y_test=train_test_split(data,target,test_size=0.25)

print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')

#25% danych testowych
````

Rysunek ze zbiorem treningowym (szary) i testowym (zloty)

````python
plt.figure(figsize=(8,6))
plt.title('Regresja liniowa train vs. test')
plt.xlabel('cecha x ')
plt.ylabel('zmienna docelowa')
plt.scatter(X_train,y_train,label='zbior treningowy',color='gray',alpha=0.5)
plt.scatter(X_test,y_test,label='zbior testowy',color='gold',alpha=0.5)
plt.legend()
plt.plot()
````

````python
#wizualizacja nam pokazuje, ze zbior treningowy i testowy zostal dobrany w sposob losowy
````

````python
#tworzymy instacje do modelu regresji liniowej 
regressor=LinearRegression()
#dopasowujemy nasz model do danych treningowych
regressor.fit(X_train,y_train)
````

````python
#oceniamy nasz model na dancyh treningowych 
#R2 score 
regressor.score(X_train,y_train)
````

````python
#teraz sprawdzmy wynik R2 score na danych testowych
regressor.score(X_test,y_test)
#ten wynik R2 score na zbiorze testowym jest dla nas wazniejszy niz na 
#zbiorze treningowym
````

Regresja liniowa-zbiÃ³r treningowy-wizualizacja

````python
plt.figure(figsize=(8,6))
plt.title('Regresja liniowa: zbior treningowy')
plt.xlabel('cecha x ')
plt.ylabel('zmienna docelowa ')
plt.scatter(X_train,y_train,label='zbior treningowy',color='gray',alpha=0.5)
plt.plot(X_train,regressor.intercept_+regressor.coef_[0]*X_train,color='red')
plt.legend()
plt.plot()
````

Regresja liniowa-zbior testowy-wizaulizacja

````python
plt.figure(figsize=(8,6))
plt.title('Regresja liniowa: zbior testowy')
plt.xlabel('cecha x ')
plt.ylabel('zmienna docelowa')
plt.scatter(X_test,y_test,label='zbior testowy',color='gold',alpha=0.5)
plt.plot(X_test,regressor.intercept_+regressor.coef_[0]*X_test,color='red')
plt.legend()
plt.plot()
````

Predykcja na podstawie modelu

uÅ¼ycie miar do oceny modelu regresyjnego 


````python
y_pred=regressor.predict(X_test)

predictions=pd.DataFrame(data={'y_true': y_test,'y_pred':y_pred})
predictions.head()


#obczajmy jak wyszlo obliczanie y_pred na podstawie oblicoznych parametrow 
# a i b i zobaczymyw  DF jak to wyglada obok y_true, ktory jest targetem
````

````python
#dodajmy do DF kolumne pokazujaca roznice miedzy y_true, a y_pred
predictions['error']=predictions['y_true']-predictions['y_pred']
predictions.head()
````
