> Funkcja *plot* jest wykorzystywana najczęściej do rysowania punktów na wykresie. Domyślnie, rysuje linię między dwoma punktami (można również rysować linie między większą ilością punktów).

```python
import matplotlib.pyplot as plt
import numpy as np

# dane do pierwszego wykresu
x1 = np.array([0, 3, 7, 9])
y1 = np.array([5, 2, 15, 7])

# rysujemy pierwszy wykres - rysunek ma jeden wiersz, dwie kolumny;
plt.subplot(1, 2,1)
plt.plot(x1,y1)
# dodajemy tytuł wykresu
plt.title("PIERWSZY")

# dane do drugiego wykresu (jeśli nie zdefinujemy wartości dla osi x, 
# domyślnie zostaną im przypisane wartości 0,1,2, ... - w zależności od ilości elementów zdefiniowanych dla y)
y2 = np.array([10, 20, 30, 40])

# rysujemy drugi wykres
plt.subplot(1, 2, 2)
plt.plot(y2) #domyslnie x=0,1,2,3
# dodajemy tytuł wykresu
plt.title("DRUGI")

plt.show()
```

## **Zadanie nr 2** - augmentacja danych

> Augmentacja danych jest jedną z możliwości, która pozwala na zapobiegnięcie nadmiernemu dopasowaniu modelu. Tą technikę można wykorzystać do sztucznego zwiększenia danych - stworzenie danych zmodyfikowanych na podstawie już istniejących.

> Do danych początkowych możemy wprowadzić różne zmiany. Dla obrazów mogą to być przekszatłcenia geometryczne lub zmiana przestrzeni kolorów, a dla danych tekstowych - tasowanie. Augmentacja daje wiele możliwości ze względu na możliwość używania wszystkich metod jednocześnie - pozwala to uzyskać wiele unikalnych próbek.

Augmentację danych można przeprowadzić na kilka sposobów, między innymi wykorzystując funkcje wbudowane w różne biblioteki. Inną możliwością jest stworzenie własnej funkcji, która z powodzeniem pozwoli na uzyskanie oczekiwanych efektów.

Poniżej przedstawiony został przykład funkcji, która bazując na możliwościach jakie daje bibliteka OpenCV (obrót obrazu o zadany kąt z wykorzystaniem *getRotationMatrix2D* oraz *warpAffine*). Jako argumenty funkcja przyjmuje: obraz, który ma zostać poddany augmentacji oraz kąt o jaki obraz ma zostać obrócony.

> [OpenCV - dokumentacja warpAffine ](https://docs.opencv.org/3.4/d4/d61/tutorial_warp_affine.html)


Tak stworzona funkcja może być jedną z wielu funkcji wchodzących w skład klasy, która będzie odpowiedzialna za augmentację danych.



```python
import cv2
def augment_rotation(img, angle=90):
    (h,w) = img.shape[:2] # wysokość i szerokość obrazu
    scale = 1.0
# macierz obrotu
    M = cv2.getRotationMatrix2D((w/2,h/2), angle, scale)
# obrót
    image = cv2.warpAffine(img,M,(w,h))
    return image
```

Sprawdźmy, jakie działanie ma stworzona powyżej funkcja, wykorzystując obraz *lab3_1.jpg*.

```python
#wczytujemy obraz
import matplotlib.pyplot as plt
obraz = cv2.imread("/content/lab3_1.jpg")
obraz_obrocony = augment_rotation(img=obraz,angle=-30)
plt.imshow(obraz_obrocony)
plt.show()
```

Jak widzimy obraz został obrócony o zadany kąt, ale również obcięty.

Wewnątrz funkcji *augment_rotation* został zdefiniowany parametr *scale*, który z powodzeniem mógłby być argumentem tej funkcji - wtedy jednocześnie można byłoby obracać i skalować obraz.

**2a**. Stwórz klasę o nazwie *ImageAugmentation*, a następnie utwórz w niej **trzy** funkcje pozwalające na augmentację danych (niech każda z funkcji ma inne zastosowanie, także inne od przedstawionego przykładu powyżej).

Może to być na przykład: odwrócenie w pionie/poziomie, przesunięcie, dodanie szumu, zmiana rozmiaru obrazu, konwersja do skali szarości, zmiana jasności lub nasycenia.

Do tworzenia funkcji wykorzystaj możliwości biblioteki OpenCV.



> [Dokumentacja - OpenCV](https://docs.opencv.org/master/d6/d00/tutorial_py_root.html)



```python
# tutaj umieść kod
class ImageAugmentation:
    def __init__(self, image):
        self.image = image
 
 ### I
    def crop(self, center, width, height):
        self.image=cv2.getRectSubPix(self.image, (width, height), center)
        return self

###II
    def rotate(self):
        self.image=cv2.rotate(self.image,cv2.ROTATE_180)
        return self

###III 
    def transopnse(self):
        self.image=cv2.transpose(self.image)
        return self

###IV
    def resize(self, dim = (100, 100)):
        self.image = cv2.resize(self.image, dim, interpolation = cv2.INTER_CUBIC)
        return self


    def show(self):
        plt.imshow(self.image)
        plt.show()
        return self
  

```

```python
def saturation_and_visualize(original):
    fig = plt.figure()
    plt.subplot(1,2,1)
    plt.title('Obraz oryginalny')
    plt.imshow(original)

    plt.subplot(1,2,2)
    plt.title('Obraz obrócony o 90 stopni')

    saturation = tf.image.rot90(original)
    plt.imshow(saturation)

image = mpimg.imread('/content/lab3_1.jpg')

saturation_and_visualize(image)
```

Wykorzystując możliwości *tf.image*, przeprowadź dwa eksperymenty z różnymi możliwościami przetwarzania obrazów (wykorzystanie minimum dwóch różnych możliwości).

> [Dokumentacja - tf.image ](https://www.tensorflow.org/api_docs/python/tf/image)

```python
#I opcja 
def image_transponse(original):
    fig = plt.figure()
    plt.subplot(1,2,1)
    plt.title('Obraz oryginalny')
    plt.imshow(original)

    plt.subplot(1,2,2)
    plt.title('Obraz transponowany')    
    f=tf.image.transpose(image)
    plt.imshow(f)

image = mpimg.imread('/content/lab3_1.jpg')

image_transponse(image)
```

```python
#II opcja 

def adjust_brightness(original):
    fig = plt.figure()
    plt.subplot(1,2,1)
    plt.title('Obraz oryginalny')
    plt.imshow(original)

    plt.subplot(1,2,2)
    plt.title('Obraz rozjaśniony')

    s = tf.image.adjust_brightness(image,delta=0.7)
    plt.imshow(s)

image = mpimg.imread('/content/lab3_1.jpg')

adjust_brightness(image)
```

**2d**. Augmentacja danych może zostać również przeprowadzona z wykorzystaniem biblioteki Keras, w której znajdują się różne warstwy przetwarzania wstępnego. 
Poniższy przykład ilustruje wykorzystanie dwóch warstw - losowego powiększenia oraz przeskalowania obrazu.

> Zbiór danych *rock_paper_scissors* zawiera prawie 3000 obrazów rąk osób grających w kamień, papier, nożyczki.



```python
from tensorflow.keras import layers
#załadowania zbioru danych i podział na zbiory - treningowy, walidacyjny i testowy
(train_leaves_ds, val_leaves_ds, test_leaves_ds), metadata = tfds.load(
    'rock_paper_scissors',
    split=['train[:85%]', 'train[85%:95%]', 'train[95%:]'],
    with_info=True,
    as_supervised=True
)

# pobranie jednego obrazu ze zbioru treningowego 
image, label = next(iter(train_leaves_ds))

#dodanie obrazu do batcha
image = tf.expand_dims(image, 0)

#docelowy rozmiar obrazu
IMG_SIZE=250

resize_zoom_and_rescale = tf.keras.Sequential([
  layers.experimental.preprocessing.Resizing(IMG_SIZE, IMG_SIZE),
  layers.experimental.preprocessing.RandomZoom(0.5),
  layers.experimental.preprocessing.Rescaling(1./255) #standaryzacja wartości pikseli 0-1
])

image_after_augmentation = resize_zoom_and_rescale(image)

plt.imshow(image_after_augmentation[0])
```
