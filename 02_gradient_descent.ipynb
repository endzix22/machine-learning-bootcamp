```python
import numpy as np
import pandas as pd
import plotly.express as px 

np.random.seed(42)
```

Wygenerowanie danych

```python
X1 = np.array([1, 2, 3, 4, 5, 6])
Y = np.array([3000, 3250, 3500, 3750, 4000, 4250])
m = len(X1)

print(f'Lata pracy: {X1}')
print(f'Wynagrodzenie: {Y}')
print(f'Liczba próbek: {m}')
```

Przygotowanie danych

```python
X1 = X1.reshape(m, 1)
Y = Y.reshape(-1, 1)
print(X1)
print(X1.shape)
```

```python
bias = np.ones((m, 1))
print(bias)
print(bias.shape)
```

```python
#ta macierz x (obczaj zeszyt), dodaje sie bias ten ala wektor w macierzy 
```

```python
X = np.append(bias, X1, axis=1)
print(X)
print(X.shape)
```

Losowa inicjalizacja parametrow

```python
eta=0.01
#wskaznik uczenia, o tyle bedzie te dalsze kroki wykonywac
weights=np.random.randn(2,1) #losowa inicjalizacja wag 
#zobacz te wagi na koncu dwie ostatnie 
print(X)
print(weights)
```

Metoda gradientu prostego

```python
intercept=[]
coef=[]
#bedziemy dodawac wartosci wo i w1 (czyli a i b)


for i in range(3000):
    gradient = (2 / m) * X.T.dot(X.dot(weights) - Y) #wzor na gradient
    weights = weights - eta * gradient 
    #nasze wagi aktualizujemy, w naszym kroku bedziemy odejmowac oblicozny gradient razy
    #wskaznik uczenia, co stanowi nowe wagi:
    
    intercept.append(weights[0][0])
    coef.append(weights[1][0])

print(weights)
```

```python
#2749.95599051->wo
#250.01027971->w1
#zatrzymalnie algorytmu po 3000 krokow. 
```

```python
df=pd.DataFrame(data={'intercept':intercept,'coef':coef})
df.head()
#utworzenie dataframe z naszych tablic 
```

Wizualizacja dopasowania

```python
px.line(df, y='intercept', width=800, title='Dopasowanie: intercept')
#wykres dla b  wo
```

```python
px.line(df, y='coef', width=800, title='Dopasowanie: coef')
#wykres dla a w1
```
